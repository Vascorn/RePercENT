```python
import os, sys
import pickle
from MultiBench.datasets.affect.get_bert_embedding import bert_version_data

dataset = 'mustard'
file_name = 'sarcasm.pkl'
root_path = os.path.abspath(os.path.join(os.getcwd(), '..'))
data_path = os.path.join(root_path, 'data', 'multibench', dataset, file_name)

with open(data_path, "rb") as f:
        alldata = pickle.load(f)

train_keys, test_keys = list(alldata['train']['id']), list(alldata['test']['id'])
text_data_path = os.path.join(root_path, 'data', 'multibench', dataset, 'sarcasm_raw_text.pkl')
new_train_data = bert_version_data(alldata['train'], text_data_path, train_keys)
new_test_data = bert_version_data(alldata['test'], text_data_path, test_keys)
```

```python
def print_data_info(data, split):
    print(f"{split} data:")
    print(f"keys in data: {list(data.keys())}")
    for key, val in data.items():
        if key != 'id':
            print(f"  {key}: {len(val)} samples")
print_data_info(new_train_data, 'Train')
print_data_info(new_test_data, 'Test')
    
```
